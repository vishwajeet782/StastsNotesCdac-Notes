**KNN CLASSIFIER:** k nearest neighbour classifier:

--> k is the no of numbers you can choose



```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.neighbors import KNeighborsClassifier as knc
```

```python
df = pd.read_excel('CDAC_DataBook.xlsx',sheet_name='iris')
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('Species',axis=1),df['Species'],test_size=0.25)
```

```python
mod2 = knc(n_neighbors=4).fit(x_train,y_train)
```

```python
y_pred2 = mod2.predict(x_test)
y_pred2
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred2))
```

compare confusion matrices of naive bayes and knn, here naive bayes showing 1 error value, but knn shows 3 error values.

Hence knn's accuracy is more. So knn is chosen among the two for this example.

**HIERARCHIAL CLUSTERING:**

The final output resembles like ORGANOGRAM which is called **DENDROGRAM**

IN this clustering, the datapoints having similar attributes will be clustered. eg:districts

AT level-2,
If you go for some more differences, then some clusters will be clustered together. eg: state

At level -3,
more differences can be clubbed together.
eg: country

at level-4
eg: continent

at level -5;
eg: world


at the lower level,the dissimilarities will be less
as the level increases, the dissimilarities will increase.

**ERROR:**

The error of an individual forecast is the difference between the actual value and the forecasted value.

**error = the actual value - the forecast value**.

**et = xt - Ft (t is subscript)**

dont calculate avg error, because, if errors are -5 and +5, then avg error is error.
in another case, the errors can be -10 and +10, the avg error=0

by this we conclude that both models or techniques are same.
Hence dont  consider calculating avg error.

-->SO, calculate
**MEAN ABSOLUTE DEVIATION**:

--> Take out the absolute value of mean deviation.

MAD = SIGMA |ei| / number of forecasts

MAD = sum of absolute value of errors / number of forecasts.

--> **MEAN SQUARE ERROR**

--> CALCULATE the squares of deviations from the mean.

MSE = sum of Squares of  absolute value of errors / number of forecasts.

MSE = sigma ei **2 /number of forecasts.

for interpretation,
we calculate,
**ROOT MEAN SQUARE ERROR**

--> RMSE which is square root of MSE.
eg: square rupees interprettion is not better so we square root it.

**MAPE:**- **MEAN ABSOLUTE PERCENTAGE ERROR**:

actual term- forecast = error
take absolute value
each error term is divided by actual term.
