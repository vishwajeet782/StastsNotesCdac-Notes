**2 SAMPLE t test**

--> 2 SAMPLE t test  is used to compare the differences of the averages of 2 independent samples with the claim value and establish whether the difference is statistically significant or not.

--> predictor is categorical, response is continuous.

```python
x = [3,7,6,5,9,8,2,3,5,4,6,7]
```

```python
y = [9,6,8,10,6,4,7,8,9,10]
```

**claim:**

H0 mean(y) - mean(x) >=2.7

Ha mean(y) - mean(x)  < 2.7 (left tail test)



```python
np.mean(y) - np.mean(x)
```

```python
statsmodels.stats.weightstats.ztest(y,x,value=2.7,alternative='smaller')
```

here p > LOS, the null hypothesis is not rejected.
The statistical difference is not significant.

the differnce between 2.28 and 2.7 can be ignored.

```python
statsmodels.stats.weightstats.ztest(y,x,value=4.2,alternative='smaller')
```

here we changed null hypothesis
H0 mean(y) - mean(x) >=4.2

Ha mean(y) - mean(x) < 4.2


here in the solution p< LOS.
THE diff between 2.28 and 4.2 cannot be ingored.
The difference is significant.
change is requied.

RIGHT TAIL TEST:

claim:

TWO TAIL TEST:

claim:

```python
statsmodels.stats.weightstats.ztest(y,x,value=1.5,alternative='two-sided')
```

here, p> LOS,

null hypothesis is not rejected.

the difference is not significant.

```python
import statsmodels.api as sm
```

```python
from statsmodels.formula.api import ols
```

```python
import statsmodels.stats.multicomp
```

Creating the model:

sample --> predictor

value --> response

output --> outcome

```python
mod1 = ols('Value ~ Sample', data = df).fit()

#creating the model
```

```python
tb1 = sm.stats.anova_lm(mod1)

#sm.stats.anova_lm is a function that computes the ANOVA table.
#The mod1 object obtained from fitting the linear regression model is passed as the argument to the anova_lm function.
#The function computes the ANOVA table, which provides information about the sum of squares, degrees of freedom, mean squares, F-statistic, and p-values for each term in the linear regression model.
#The ANOVA table is then assigned to the variable tb1.
```

```python
tb1

# here the samples are A,B,C
#F stands for Fisher value, this is used in ANOVA calculations
#df - degree of freedom (sample size -1 =2)
#residual = no of rows - no of samples.
```
