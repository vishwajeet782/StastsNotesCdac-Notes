```python
import pandas as pd
```

```python
df1 = pd.read_excel('locations.xlsx',sheet_name = 'Pune')
```

```python
df2 = df1 = pd.read_excel('locations.xlsx',sheet_name='Mumbai')
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2, on = 'Subject', how = 'inner')

#on = common columns
#how = what kind of merging you want
```

```python
pd.merge(df1,df2, on = 'Subject', how = 'outer')

#OUTER MERGE
```

```python
pd.merge(df1,df2, on = 'Subject', how = 'left')

#LEFT MERGE
#the first dataframe(or left df) will be given more priority
```

```python
pd.merge(df1,df2, on = 'Subject', how = 'right')

#RIGHT MERGE
#the second dataframe(or right df) will be given more priority
```

```python
pd.merge(df1,df2,on=['Subject','Grade'],how='inner')

#both subject and grade must be same or common in inner
```

```python
pd.merge(df1,df2,on=['Subject','Grade'],how='outer')
```
