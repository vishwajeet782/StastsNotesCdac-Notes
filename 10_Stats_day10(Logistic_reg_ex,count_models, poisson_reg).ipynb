**VARIANCE INFLATION FACTOR(VIF):**
A variance inflation factor (VIF) is a measure of the amount of multicollinearity in regression analysis. Multicollinearity exists when there is a correlation between multiple independent variables in a multiple regression model. This can adversely affect the regression results. Thus, the variance inflation factor can estimate how much the variance of a regression coefficient is inflated due to multicollinearity..

```python
df = pd.read_excel('CDAC_DataBook.xlsx',sheet_name='poisson')
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('num_awards',axis=1),df['num_awards'],test_size=0.25)
```

```python
x_train = sm.add_constant(x_train,prepend=False)
```

```python
from statsmodels.discrete.discrete_model import Poisson as psn
```

here, x_train and y_train has to be clubbed together.

```python
df_train = pd.concat([x_train,y_train],axis=1)
```

```python
df_train.head()

#here 2 and 3 are prog2 and prog3
```

```python
df_train.columns
```

```python
df_train.rename(columns = {2:'prog2',3:'prog3'},inplace=True)
```

```python
df_train.head()
```

```python
mod1 = psn.from_formula('num_awards ~ math + prog2 + prog3', data=df_train).fit()
```

```python
print(mod1.summary())
```

prog1 is taken as reference event,
prog2 and prog3  values are with reference to prog1


**H0 for math:**
--> the math score does not influence num_awards.

here p<0.05 ,H0 is rejected. hence, math marks influence the num awards(no of job offers).

--> for every unit increase in math marks, the job offers will increase by 0.0735

**H0 for prog2:**

--> the movement from prog1 to prog2 the no of job offers remain the same.
--> as p<0.05, **H0 is rejected**.
--> so when we move from prog1 to prog2 the job offers are changing.
when we move from prog1 to prog2, the job offers increase because coeff of prog2 is 1.0062.

**H0 for prog3:**
the movement from prog1 to prog3 the no of job offers remain the same.
as p>0.05, H0is not rejected.
so, if you join in prog1 or prog3 the job offers will remain same.

So, by this two cases, if you join prog2 the job offers will increase.
