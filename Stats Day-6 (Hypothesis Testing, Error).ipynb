**HYPOTHESIS TESTING:**

**Steps involved in Hypothesis Testing:**

**Step - 1:** (very very imp)

--> **Establish the null and alternative hypothesis.**
--> H0 = null hypothesis, H1/HA = alternative hypothesis
--> In statistics, the null hypothesis (Hâ‚€) is a statement that assumes there is no significant difference or relationship between variables being studied
--> Null Hypothesis is made in such a way that if it is found to be true
      - No action to be taken
      - No decision
      - No changes has to be made.

**Step -2:**

--> Establish the confidence level and the level of significance for the analysis.
--> confidence level default value = 95%

-->  **level of significance:**
  --> **LOS is the max allowed probability of committing type - I error.**
     - LOS = 100 - Confidence level
     - or in fraction LOS = 1 - Confidence level
     - default LOS = 5%



**Step -3:**

--> **Choose the appropriate statistical test for performing the analysis**.

**Step -4:**

--> Use the appropriate functions in python to calculate the p-value.

--> **p - value = Actual probability of committing type-I error.**

                OR

  **p-value **= If the Sample Statistic (SS) and Population Parameter(PP) or the claimed value do not meet the desired criteria then, p-value gives the probability that the difference is just a matter of chance, there is no actual difference and so the action can be avoided.

**Step - 5:**

--> Compare the p-value (in step-4) with the LOS(level of significance)

--> **If p-value is lessthan LOS, then we reject the Null Hypotheis**.
(V.V.V imp****)

--> If p-value is more than LOS, then we do not reject( remember! there is a huge difference between not rejecting and accepting) the Null Hypotheis

Now the bias is removed.

--> SS = PP + Actual Error + Chance Variation.

**Errors:**
Threre are two errors

**1) Type - I error:**

--> Type I error occurs when we Reject a null hypothesis (H0) which is true.

--> that is rejected due to limited information because we're working with a sample.

example 1:
-->Not watching a movie which is good which is not done a good/fancy marketing.

example: rejecting a girl in arrange marriage.

**2) Type - II error:**

--> Not Rejecting a null hypothesis (H0) which is false.

example 1:
--> Watching a movie which is not good.

--> **Inorder to decrease both the errors, increase the confidence level which is done by taking the whole data i.e., population which is not possible.**

Example:

In covid doctor says if the prob of catching infrction is less than 7% you can go out.

Max probability of catching infection = 7% = LOS

H0 = stay at home
HA = move out of home.

case 1: if one day the prob is 23% i.e., p value = 23% then,
we stay at home.
--> p-value > LOS
in this case we're accepting the Null Hypothesis.

case 2: if p = 3%, then we'll go out.
--> P-value < LOS
here, we're rejecting the Null Hypothesis.

1) **1 Sample Z test or 1 Sample t test:**

Step 1:

H0 - avg(x) is atleast 7.2 ;
H0: a >= 7.2
HA: a < 7.2

Step 2:

confidence level = 95%, LOS = 5% = 0.05

Step 3:

choose 1 sample t test or z test to compare a sample mean with the claim value

Step4: use appropriate function

syntax:
**statsmodels.stats.weightstats.ztest(array,value = claim_value, alternative='smaller' or 'larger',)**


**output will be:**

**(Z or Test statistic, p - value) **


```python
statsmodels.stats.weightstats.ztest(x,value=7.2,alternative='smaller')

#outut is (z or test statistic,  p- value)
#alternative ='smaller' because there is lessthan symbol in  alternative hypotheis.
```

**Step 5:**

**compare p-value with LOS**

Here, p-value < LOS i.e., 0.025 < 0.05

hence, we reject null hypothesis(H0).
The difference between 6 and 7.2 is statistically significant, cannot be ignored so the Action is required.


Changing the Null hypothesis H0:

H0 avg(x) is atleast 6.5
H0 a >= 6.5, HA a < 6.5


```python
statsmodels.stats.weightstats.ztest(x,value=6.5,alternative='smaller')
```

p-value is morethan 0.05 so we DO NOT reject the H0. The difference between 6 and 6.5 is statistically NOT significant, can be ignored so the action is not required.

**V.V.IMP**


--> if p < LOS (0.05), that means the diff between SS and PP is large enough not to be ignored, So action is required.

--> The diff is not only due to chance variation,bias. there is an actual difference.
