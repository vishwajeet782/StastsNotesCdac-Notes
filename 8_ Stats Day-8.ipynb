```python
import statsmodels
from statsmodels import stats
import statsmodels.api as sm
```

```python
mydf = df[['rank','gender','salary']]
```

```python
mydf.head()

#rank and gender are the predictors.salary is the response or output.
```

```python
mod1 = ols('salary~rank+gender', data=mydf).fit()
```

```python
tb1 = sm.stats.anova_lm(mod1)

```

```python
print(tb1)
```

establish null hypothesis:

--> H0 for rank:
H0: The average salary is same for all the pairs of ranks.

--> H0 for gender:
H0: The average salary is same for all the pairs of genders.

**In general:**
H0: The average **response** is same for all the pairs of **predictor**.

**For rank:** since the p-value(1.234599e-43) is lessthan 0.05, so we reject the H0. Salary depends on rank.

**For gender:** since the p-value(2.202874e-01) is morethan 0.05, so we DO NOT reject the H0. Salary does not depends on gender.

we're considered both rank and salary, dependency on rank is more and comparing to rank there is negligible dependecy on the gender.

**2 PROPORTION TEST**

In this test, the difference of proportions of two samples is calculated and compared against the claim value and establish whether the difference is statistically significant or not.

--> My claim is that Banglore has atleast 12% more enginners than Kharghar.

H0= prop(BLR) - prop(KHR) >= 0.12

HA = prop(BLR) - prop(KHR) < 0.12

In BLR, in a sample of 150 students, 125 were engineers

In KHR, in a sample of 100 students, 75 were engineers.



```python
125/150 - 75/100
```

--> We need to check whether the difference between 0.083 and 0.12 is statistically significant or not.

--> Here we create two arrays

```python
ss = np.array([150,100])

#sample space
```

```python
en = np.array([125,75])

#no of engineers
```

```python
statsmodels.stats.proportion.proportions_ztest(en,ss,0.12,alternative = 'smaller')

#claim value= 0.12
```

--> p value = 0.24 > 0.05,

--> so the H0 is not rejected

--> the difference between 0.12 and 0.083 is not significant

--> the difference is due to chance variation
