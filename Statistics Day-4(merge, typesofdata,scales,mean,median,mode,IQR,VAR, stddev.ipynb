**STATISTICS THEORY:**

There are three types of data in statistics point of view:
1)continuous data
2)Discrete data
3)Attribute data

***Continuous and Discrete data will always be numeric***

**1)continuous data:**
--> this is the data of numbers which have any no of decimal points are possible.
--> eg: temp of mumbai
    weight
eg2: time the local train takes to reach from kharghar to cst

**All measurements will be continuous data.**



**2)Discrete data**

--> The data where the numbers cannot have a decimal point.


eg: no of students in class(84.3 students are not possible, it's either 84 or 85)

-> this data is the result of **counting.**
-> this data is always a whole number

eg) no of stations between kharghar and cst
  no of countries
  no of chairs



**3)Attribute data**

--> This data is **characteristic.**
--> Attribute data can be a number or can be string.
--> If the data is string, then understand it is attribute data

eg: some use colab, some use spyder (counting according to the attribute)
colab = 15
spyder = 10

eg:2)
giving rating of ipl on a scale of 1-5(numbers as the attibutes have some restrictions) or A-E
but the count of no of persons who gave the particular rating is discrete data.

eg.3) Age is continuous but gender is attribute data.

drinking water(continuous data) from where you brought the water = canteen(attribute data) or colour of the bottle is blue(attribute data).

eg4) the type of class in which a person travels in the local train is attribute data. eg: normal,ac.



There are four types of scales in statistics:

**1)Nominal Scale:**

--> **The scale where we have multiple categories, the sequence of categories does n't matter.** or **Categories without logical order**

--> if there is an attribute data, that can be on Nominal scale

eg: name the languages you know. you'll speak hindi lang even if you give hindi any position in the list of languages you know


**2) Ordinal scale:**

--> **Categories with logical order**

eg: ranking (it is given on the order that 1-good, 5 bad (or) 1-bad, 5-good)

--> the order can change. but it will be in a logical order like 1-5 or 5-1.

--> The data is attribute data here

eg: give the rating of how much delay the mumbai local makes.(1 bad-5 very good)

**3)Interval scale:**

--> The scale where the number can be in decimals
--> **zero** does not mean absence of measurement.
we don't use this often

eg: temperature measurement.

**4) Ratio scale:**

-->We can perform all kind of measurements on this scale.
-->All the continuous and discrete data will be on ratio scale.
-->comparisions.
eg:harbour line trins are twice as delayed as central line trains.
eg: comparision of height, weight, experience etc

```python
freq*values/sum(freq)
```

When to use  mean and median? #IMP EXAMPLE



```python
#eg:suppose the salaries in placements in a college are:

sal = [6,7,6,7,8,7,6,7,8,7,6,7,8,8,7,6,7,6,7,7,6,8]
```

```python
np.mean(sal)
```

```python
np.median(sal)
```

```python
new_sal = [6,7,6,7,8,7,6,7,8,7,6,7,8,8,7,6,7,6,7,7,6,8,54,50,52]
```

```python
np.mean(new_sal)
```

```python
np.median(new_sal)

# This figure is more close to average salary.
#Mathematically we should never use mean under any circumstance always use median.
#mean always misleads the decision.
```

Mean should be used when:

1)Data distribution is symmetric

2)Data should not have any extreme values(values that are far greater than the normal range. see the example above)

***If the data is symmetric then mean will be equal to median***
